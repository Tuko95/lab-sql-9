#### Write a function that checks if customer borrowed more or less films in the month of June as compared to May

```python
# Joining both dataframes

total_rental = rentals_may.join(rentals_june, lsuffix='_may', rsuffix='_june')

total_rental = total_rental.drop('customer_id_june', axis=1)

total_rental.rename(columns = {'customer_id_may':'customer_id'}, inplace = True)

display(total_rental)
```

```python
def month_comparator(item):
    for item in total_rental['customer_id']:
        if (total_rental['Nº rental_id_may'] > total_rental['Nº rental_id_june']):
            print('The customer borrowed more films in May')
            break
        elif total_rental['Nº rental_id_may'] == total_rental['Nº rental_id_june']:
            print('The customer borrowed the same number of films both months')
            break
        else:
            print('The customer borrowed more films in June')
            break
```
